## Remoção de outliers

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #distancia interquartil
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

```python
for base in listabases:
    base = remove_outlier(base, "value")
    dataframe = pd.concat([dataframe, base], ignore_index=True)
```

Aqui há uma contradição entre as visualizações e a matriz de correlação.
Nas visualizações, vemos cidades se repetirem entre as que mais emitem, enquanto, na matriz de correlação, vemos um baixo valor absoluto para correlação entre cidade e concentração de poluentes. Daí, vê-se a importância das visualizações!


Talvez isso tenha ocorrido, porque, diferentemente da concentração de poluentes, as variáveis fictícias atribuídas às cidades não estabelecem uma autocorrelação linear.


Um possível forma de resolver isso, é lançando mão da correlação espacial, porém isso foge do escopo deste projeto.
